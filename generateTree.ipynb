A helper file to generate a random Tree.
Calculations in the lower part of the file are just for checking whether the tree created is a valid one

```python
import sys, os
sys.path.append(os.path.abspath('..'))
from graphilp.imports import ilpgraph, graph_formats
from graphilp.imports import networkx as imp_nx
import stp_flow
import networkx as nx
from gurobipy import *
import xlsxwriter
import random
import matplotlib.pyplot as plt
import pcst_fast
%load_ext autoreload
%autoreload 2
```

```python
if __name__ == '__main__':
    path = os.path.join(os.getcwd(), "bip42u.stp")
    G, terminals = graph_formats.stp_to_networkx(path)
    
    prizes = []
    nodes = []
    edges = []
    costs = []
    i = 0
    root = 1200
    
    for node in range(len(G.nodes())):
        if node in terminals:
            prizes.append(1000000)
        else:
            prizes.append(0)
        nodes.append(node)
    for edge in G.edges():
        edges.append([edge[0] - 1, edge[1] - 1])
        costs.append(1)
        
    verticesSol, edgesSol = pcst_fast.pcst_fast(edges, prizes, costs, root, 1, 'none', 1)
    
    edgesSol = sorted(edgesSol)
    for edge in G.edges(data=True):
        if i in edgesSol:
            edge[2]['Capacity'] = 2
        else:
            edge[2]['Capacity'] = 0
        i += 1
    
    Graph = imp_nx.read(G)
    model = stp_flow.create_model(Graph, terminals, 'weight', 0, root)
    model.Params.Threads = 3
    model.optimize() # stp_flow_cut.cut_cycle
    
    
```
